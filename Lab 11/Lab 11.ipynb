#### Task 1 - Read the image ‘eight.tif’. Write a function named ‘myMorphology’ to extract the boundaries of coins from the read image. 
[HINTS]:
1. First close the image then perform erosion.
2. Take the difference of two images to find boundari


es.


#### Loading and Displaying the Image

```python
image = cv2.imread('image.png', 0)
plt.figure(figsize=(5,5))
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')
plt.show()
```

#### Applying the morphological operation Closing on the image

```python
kernel = np.ones((5, 5), np.uint8)
closed_image = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')

plt.subplot(1,2,2)
plt.title('Closed Image')
plt.imshow(closed_image, cmap = 'gray')
plt.show()
```

#### Applying Erosion

```python
eroded_image = cv2.erode(closed_image, kernel, iterations=1)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')

plt.subplot(1,3,2)
plt.title('Closed Image')
plt.imshow(closed_image, cmap = 'gray')

plt.subplot(1,3,3)
plt.title('Eroded Image')
plt.imshow(eroded_image, cmap = 'gray')

plt.show()
```

#### Subtracting the eroded image from the closed image to obtain boundaries

```python
boundary_image = closed_image - eroded_image
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,4,1)
plt.title('Original Image')
plt.imshow(image, cmap = 'gray')

plt.subplot(1,4,2)
plt.title('Closed Image')
plt.imshow(closed_image, cmap = 'gray')

plt.subplot(1,4,3)
plt.title('Eroded Image')
plt.imshow(eroded_image, cmap = 'gray')

plt.subplot(1,4,4)
plt.title('Extracting Boundaries')
plt.imshow(boundary_image, cmap = 'gray')

plt.show()
```

#### 7x3 Kernel and 11x3 Kernel

```python
opened_image_1 = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel_1)
opened_image_2 = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel_2)
```
