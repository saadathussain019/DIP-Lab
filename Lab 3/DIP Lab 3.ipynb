#### Displaying Binary Image  

```python
# Load image in grayscale
original_image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

# Get image dimensions
height, width = original_image.shape

# Create empty binary image
binary_image = np.zeros((height, width), dtype=np.uint8)

# Set threshold
my_threshold = 127

# Loop through image and binarize
for y in range(height):
    for x in range(width):
        if original_image[y, x] > my_threshold:
            binary_image[y, x] = 255
        else:
            binary_image[y, x] = 0

# Display original and binary images
cv2.imshow('Original Image', original_image)
cv2.imshow('My Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Displaying Results of Manual Binarization

```python
# Open webcam
my_webcam = cv2.VideoCapture(0)

while True:
    # Read frame from webcam
    ret, frame = my_webcam.read()
    
    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Display grayscale frame
    cv2.imshow('Grayscale Webcam', gray_frame)
    
    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release webcam and close windows
my_webcam.release()
cv2.destroyAllWindows()
```
