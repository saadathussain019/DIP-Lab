#### Importing Dependencies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Loading and Displaying the Image

```python
path = 'gtr.jpg'
image = cv2.imread(path)

# Convert the image from BGR (OpenCV format) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Convert the image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

brightness_increase = 70 
image_brightened = cv2.add(image_gray, brightness_increase)

# Display the original and the brightened grayscale image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axis

plt.subplot(1, 2, 2)
plt.imshow(image_brightened, cmap='gray')
plt.title('Brightened Grayscale Image')
plt.axis('off')  # Hide axis

plt.show()
```

```python
# Contrast adjustment parameters
alpha = 2.5  # Contrast control (1.0-3.0)
beta = 0     # Brightness control (0 for no change)

# Manually adjust contrast
# Create an empty image to hold the contrast-adjusted values
image_contrast = np.zeros(image_gray.shape, dtype=np.uint8)

for i in range(image_gray.shape[0]):  # Iterate over rows
    for j in range(image_gray.shape[1]):  # Iterate over columns
        new_pixel = alpha * image_gray[i, j] + beta
        new_pixel = np.clip(new_pixel, 0, 255)
        image_contrast[i, j] = new_pixel

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')  # Hide axis

plt.subplot(1, 2, 2)
plt.imshow(image_contrast, cmap='gray')
plt.title('Manually Contrast Adjusted Image')
plt.axis('off')  # Hide axis

plt.show()
```
