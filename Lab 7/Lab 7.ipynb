### Importing Dependencies

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

#### Loading the Image

```python
image_path = "C:/Users/user/Jupyter Projects/DIP Lab/Lab 7/images/kidney.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
image.shape
```

#### Task 1 - Gray Level Slicing

#### Normalizing Pixel Values

```python
normalized_image = image/255
```

```python
#Initializing gray sliced image
gray_sliced = np.zeros_like(normalized_image)
gray_sliced.shape
```

#### Performing Gray Slicing using Nested Loops

```python
rows, columns = image.shape[:2]
print(rows,columns)
```

```python
for i in range(rows):
    for j in range(columns):
        if (normalized_image[i,j] <= 0.25):
            gray_sliced[i,j] = 1 - normalized_image[i,j] #invert for values less than 0.25
        elif (normalized_image[i,j] <= 0.5):
            gray_sliced[i,j] = 1.0 #for values between 0.25 and 0.5
        else:
            gray_sliced[i,j] = normalized_image[i,j] #original value is retained
```

```python
gray_sliced = (gray_sliced *255).astype(np.uint8)
```

#### Display the Original and Gray Sliced Images

```python
plt.figure(figsize=(10,5))
# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")

# Display the gray level sliced image
plt.subplot(1, 2, 2)
plt.imshow(gray_sliced, cmap='gray')
plt.title("Gray Sliced Image")

plt.tight_layout()
plt.show()
```

#### Task 2 - Histogram Equalization

#### Calculating the Histogram

```python
hist = np.zeros(256, dtype=int)
for row in range(rows):
    for col in range(columns):
        hist[image[row, col]] += 1
```

#### Computing Cumulative Histogram

```python
cumulative_hist = np.zeros(256, dtype=float)
cumulative_hist[0] = hist[0]
for i in range(1, 256):
    cumulative_hist[i] = cumulative_hist[i - 1] + hist[i]
```

#### Mapping to new values based on cumulative histogram

```python
output = np.zeros((rows,columns), dtype=float)
for row in range(rows):
    for cols in range(columns):
        output[row, cols] = cumulative_hist[image[row, cols]] / (rows * columns)
output = (output * 255).astype(np.uint8)
```

#### Displaying Results

```python
# Display results
plt.figure(figsize=(16, 10))
    
# Original Image
plt.subplot(2, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')    

# Original Image's Histogram
plt.subplot(2, 3, 2)
plt.hist(image.flatten(), bins=256, range=[0, 255], color='black')
plt.title('Original Histogram')

# Cumulative Histogram
plt.subplot(2, 3, 3)
plt.plot(cumulative_hist, color='black')
plt.title('Cumulative Histogram')
    
# Post Histogram Equalization
plt.subplot(2, 3, 4)
plt.imshow(output, cmap='gray')
plt.title('Post Histogram Equalization')
plt.axis('off')
    
# Equalized Histogram
plt.subplot(2, 3, 5)
plt.hist(output.flatten(), bins=256, range=[0, 255], color='blue')
plt.title('Equalized Histogram')
    
plt.tight_layout()
plt.show()
```

#### Task 3 - Contrast Stretching

```python
float_image = image.astype(np.float32)
minimum_value = np.percentile(float_image,5)
maximum_value = np.percentile(float_image, 99)

contrast_stretched_image = (255 * (float_image - minimum_value) / (maximum_value - minimum_value)).astype(np.uint8)
```

#### Displaying Results

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(contrast_stretched_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB
plt.title('Contrast Stretched Image')
plt.axis('off')

plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))

plt.subplot(1, 2, 1)
plt.hist(image.ravel(), bins=256, color='black', alpha=0.5)
plt.title('Histogram of Original Image')

plt.subplot(1, 2, 2)
plt.hist(contrast_stretched_image.ravel(), bins=256, color='black', alpha=0.5)
plt.title('Histogram of Stretched Image')

plt.show()
```
